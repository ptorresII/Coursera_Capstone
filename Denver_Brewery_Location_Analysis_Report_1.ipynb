### 1. Introduction

##### 1.1 Background

The first Colorado beer brewery opened in 1859; and famously the Coors brewery in Golden, Colorado was established in 1873. By 1910 Denver had more than 400 saloons that consumed mostly locally brewed beer; 90 were in a four-block radius. Today drinking beer in Denver is a top priority for locals and visitors alike. Denver has a rich history and array of different breweries. These Microbreweries and craft breweries are part of what makes Denver special, they are woven into the fabric of the great state of Colorado. In 2012 total business as well as individual tax revenue generated by breweries and related industries in Colorado was more than 2.5 billion. In the beginning of 2014, there were over 2800 breweries in the U.S this includes brewpubs, microbreweries, and regional craft breweries. The industry at this time employed around 43,000 Americans and had an economic impact of more than 246 billion.

In 2017 Colorado had six breweries for every hundred thousand residents. According to alcohol.org; in 2019 the state was the third largest producer of craft beers in the country making over 1.5 million barrels annually. According to Datafinity Colorado independently developed “one of the most distinct state beer cultures”. There’s a love of brewing in Colorado than is strong and vibrant and beer lovers are thirsty for something new. Colorado loves beer so much that there is a brewery at the Denver international Airport, festivals, and brewery tours throughout the city that take place daily. The Denver brewery scene is not only a great industry success but a way of life for the state of Colorado.

##### 1.2 Problem

An individual or a group may wish to open a beer microbrewery or brewpub in the city of Denver. The city is the Colorado state capital in the U.S. The state of Colorado is the 3rd largest producer of beer in the U.S.; and Denver is 8th for brewery’s per capita. Denver is a great location for a brewery, but Denver is a large city that is 155 mi². This projects goal is to help find potential geolocations to build a new brewery for the greatest chance of success. A business’s first physical location is one of the most important decisions to make beyond deciding to go into business in the first place. It’s also potentially the most expensive decision; that has the longest impact over time to a business’s success.
