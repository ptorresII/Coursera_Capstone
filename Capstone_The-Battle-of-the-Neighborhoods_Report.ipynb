### 6. Analysis

#### 6.1 One hot encoding venue category

One-hot encoding is a way to take a count of existing as well as their absences. These counts will be used to find the top five and top 10 venues located near the breweries in the data set. These counts will use created a new data frame that will be combined brewery data set for analysis. This is done by creating dummy values to indicate zero when no venue of the category is returned with one. Because these categories are encoded in a binary sequence the count of existing vendor categories can be made as well as a count of no vendor in the specific category for that brewery location.

#### 6.2 Group rows by mean frequency of occurrence

The one-hot encoding venue categories will then be grouped by a brewery. And each vendor category will have its mean frequency calculated and returned. This process essentially adds the rows up under each brewery name therefore it eliminates many rows in this data set. Because there are now missing data rows the index will now have to be reset for it to be useful in the next section. This new Denver grouped data frame can now be used for the clustering analysis.

#### 6.5 K-Means for clustering Breweries in Denver by vendor categories

For this project was important to segment the breweries into general categories with similar nearby vendors. K-Means is a commonly used clustering algorithm that is unsupervised due to no advanced information. K-Means separates data into groups or clusters that are more like each other than they are to members of other clusters. Once the individual breweries are assigned to one of the clusters a general statement about the brewery’s similarities can be made. 

*K=4 was determined by visually looking at the map of Denver with the overlaid breweries and manually visualizing apparent clusters as best practice method rather than an elbow as the elbow method cannot determine the best K in this instance. Each brewery is assigned to one of the four clusters
